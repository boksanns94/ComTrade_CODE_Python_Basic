## 13.1 Tuple

- Tuple je kolekcija objekata koja je uredjena i immutable

- Tuple je iterabla, poput liste

- Tuple objekti umesto uglastih koriste obicne zagrade (ili, ako su elementi navedeni redom i odvojeni zarezima, Python po default-u stvara tuple objekat od tih elemenata)

- Tuple se moze citati, indeksirati, slajsovati, ali se ne moze menjati

````
tup1 = ("ja", "sam", "tuple")
print(type(tup1))

print(tup1[0:2])

tup1[0] = "bio"
````

````
prazan = ()     # Prazan tuple

jednoclan = (5,)    # Mora imati zarez, jer je to sintaksna indikacija da se radi o tuple-u,
                    # ne o obicnim algebarskim zagradama

tup2 = 1, 2, 3, 4, 5   # Python default-uje kreiranjem tuple-a ovde, sto treba imati na umu
print(type(tup2))
````

````
# Tuplovi se ne mogu menjati, ali se mogu konkatenirati 
novi = tup1 + jednoclan

print(novi)
````

## 13.2 Set

- Skup je u matematici najopstija algebarska strukutra koja se ne definise - zato sto se smatra da je intuitivno jasno sta je to skup, u najopstijem smislu

- U Python-u, skup je neuredjena kolekcija elemenata bez ponavljanja - kao lista u kojoj nema nikakvog narocitog redosleda medju elementima i u kojoj je svaki element jedinstven

- Skupovi su mutable strukture - mogu se dodavati elementi u njih i brisati elementi iz njih

- Ali u skupu mogu postojati samo immutable objekti (int, float, string, tuple) - nikako liste, ili recnici

- Skupovi su veoma zgodna struktura za mnoge probleme u programiranju i cesto se koriste

````
skup = set()    # Prazan skup se ovako definise, nikako preko (), to je tuple
print(type(skup))

skup1 = set([1, 2, 3, 4, 4, 4, 5])  # Skup pravimo od liste elemenata
print(skup1)    # Obratiti paznju da je samo jedna cetvorka zavrsila u skupu

skup2 = set([1, 2, [3, 4]])     # Skup ne moze sadrzati mutable objekte
````

- Set jeste mutable tip, ali je pitanje kako iz njega dohvatati elemente ako se ne moze indeksirati

- Lako mozemo dodati nove elemente u skup

````
skup1.add(10)
skup1.update([100, 200, 300, 300])   # update se koristi kada se dodaje vise elemenata
print(skup1)
````

- Elemente mozemo brisati koristeci **discard** ili **remove** metode, od kojih obe brisu element, s tim sto druga javlja gresku ukoliko element ne postoji u skupu

````
skup1.discard(10)
skup1.discard(500)
skup1.remove(500)
````

- Takodje, iz skupa se moze **pop**-ovati element, ali posto je skup neuredjen, nema nacina da znamo koji element ce biti pop-ovan

- Skupovi podrzavaju i neke operacije preuzete iz matematike

````
s1 = set([1, 2, 3, 4])
s2 = set([4, 5, 6, 7])

# Unija ova dva skupa
print(s1 | s2)  # s1 OR s2

# Presek
print(s1 & s2)  # s1 AND s2

# Razlika
print(s1 - s2)  # Elementi koji su u s1, a nisu u s2
print(s2 - s1)  # Elementi koji su u s2, a nisu u s1

# ...
````

- Kroz skupove se moze iterirati, ispitivati da li je element u skupu, i jos operacija slicnih listama

## 13.3 Dictionary

- Recnik je kolekcija elemenata **key: value** parova

- Elementi su neke vrednosti kojima su pridruzeni neki kljucevi koji te vrednosti identifikuju

- Kljucevi moraju biti jedinstveni

- Dosta su interesantante strukture podataka i imaju veliku primenu

````
prazan = dict() # ili {}

imenik = {
    'Pera': '061123',
    'Zika': '062123',
    'Mika': '063123'
}

print(f'Zikin telefon: {imenik["Zika"]}')   # Zgodno, zar ne? Pretrazim ceo imenik po kljucu koji je 
                                            # necije ime

# Sta je jos moguce?
print(list(imenik)) # Vrati listu napravljenu od recnika

del imenik["Mika"]  # Obrisi Miku iz imenika

print("Pera" in imenik) # Je li Pera u imeniku?
print("Joka" in imenik) # Joka?

# Recnik se moze formirati i kastovanjem liste parova

lista = [("Pera", "Beograd"), ("Zika", "Nis")]
recnik = dict(lista)
print(recnik)
````

- Kroz recnike se iterira po key-value parovima pozivanjem **items()** metoda nad recnikom koji vraca upakovan tuple objekat

````
print(imenik)   # Izbacili smo Miku iz imenika malopre
imenik["Mika"] = "063123"   # Vracamo ga
print(imenik)

print()

# Iteracije
for k, v in imenik.items(): # [(key, value), (key, value)...]
    print(k, v)
````
