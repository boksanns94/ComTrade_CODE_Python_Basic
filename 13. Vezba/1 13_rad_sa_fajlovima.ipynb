### 2
Napisati program koji iz datoteke razno.txt u datoteku palindromi.txt prepisuje sve palindrome. Reč je palindrom ako se isto čita sa leve i sa desne strane bez obzira
na veličinu slova. U slučaju greške, na standardni izlaz za greške ispisati odgovarajuću poruku.

````
# Resenje
ulaz = "razno.txt"
izlaz = "palindromi.txt"

def palindrom(rec):
    if rec == rec[::-1]:
        return True
    else:
        return False

with open(ulaz, 'r') as ulazniFajl, open(izlaz, 'w') as izlazniFajl:
    reci = ulazniFajl.readlines()
    
    for rec in reci:
        rec = rec.replace("\n", "")
        if palindrom(rec):
            izlazniFajl.write(rec)
            izlazniFajl.write("\n")
````

### 6 
Napisati program koji prebrojava koliko se puta linija datoteke ulaz.txt završava stringom s koji se učitava sa standardnog ulaza. U slučaju greške na standardni izlaz za greške ispisati odgovarajuću poruku.

```
# ulaz.txt
/var/log/apache2/error.log
/var/log/dpkg.log
moj_log.log
/home/korisnik.login
/home/korisnik.log.txt

===========================
# Unos
s = ".log"
===========================
# Izlaz
Broj linija: 3
```

````
# Resenje
def endsWith(line, terminator):
    br_kar = len(terminator)
    zavrsetak = line[-br_kar:]
    
    if zavrsetak == terminator:
        return True
    else:
        return False

terminator = '.log'
with open("ulaz.txt", 'r') as fajl:
    br = 0
    for line in fajl:
        line = line.replace("\n", "")
        if endsWith(line, terminator):
            br += 1

    print(br)
````

Napisati program koji poredi dve datoteke i ispisuje redni broj linija u kojima se datoteke razlikuju. Imena datoteka se zadaju kao unos korisnika. U slučaju greške, na standardni izlaz za greške ispisati odgovarajuću poruku.

Screenshot 2020-11-24 181257.png

````
# Resenje
def uporedi(s1, s2):
    l1 = s1.split('\n')
    l2 = s2.split('\n')
    razlika = abs(len(l1)-len(l2))

    for i in range(min(len(l1), len(l2))):
        if l1[i] != l2[i]:
            print(i+1)

    if razlika != 0:
        for j in range(razlika+1, max(len(l1), len(l2))):
            print(j+1)

with open("prva.txt", 'r') as f1, open("druga.txt", 'r') as f2:
    uporedi(f1.read(), f2.read())
````

8. Napisati program koji kao korisnikov unos dobija ime datoteke koja sadrži samo brojeve, pročitati datoteku i sve brojeve učitati u listu. Napisati funkciju koja izbacuje duplikate iz učitane liste, upisati novu listu (bez duplikata) u datoteku bez_duplikata.txt

````
def duplikati(lista):
    bez_duplikata = []
    for broj in lista:
        if broj not in bez_duplikata:
            bez_duplikata.append(broj)

    return bez_duplikata

f = open("brojevi.txt", "r")
sadrzaj = f.read()
f.close()

lista = sadrzaj.split(", ")

bez_duplikata = duplikati(lista)

f = open("bez_duplikata.txt", "w")
for element in bez_duplikata:
    f.write(element)
    f.write(", ")
f.close()
````

````
def da_li_je_palindrom(rec):
    return rec == rec[::-1]

f = open("razno.txt", "r")
reci = f.readlines()
f.close()

f = open("palindromi.txt", "w")

for rec in reci:
    rec = rec.replace("\n", "")
    if da_li_je_palindrom(rec):
        f.write(rec+"\n")

f.close()
````

````
ime = "druga.txt"

f = open(ime, "r")
sadrzaj = f.readlines()
f.close()

for i in range(0, len(sadrzaj)):
    sadrzaj[i] = sadrzaj[i].replace("\n", "")

max_duzina = 0
for linija in sadrzaj:
    if len(linija) > max_duzina:
        max_duzina = len(linija)

najduze_linije = []
for linija in sadrzaj:
    if len(linija) == max_duzina:
        najduze_linije.append(linija)

print(sorted(najduze_linije)[0])
````
