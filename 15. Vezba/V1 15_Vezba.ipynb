```python
"""
ULAZ:
31;Stalingrad;Action|Drama|War|IMAX;14.03.2015;Zack Illsley;$38527192.59
109;Ghajini;Action|Romance|Thriller;29.05.2012;Zack Illsley;$96531192.13
110;Dog Tags;Drama|Romance;09.11.2012;Faustina Malkinson;$30866186.69
209;Man, Woman and the Wall (Kikareta onna no mirareta yoru);Drama|Romance|Thriller;12.07.2010;Zack Illsley;$62739931.28
330;Dead of Night;Horror|Mystery;23.01.2007;Zack Illsley;$51519915.39

IZLAZ:


Directors = {
  reziser1 : {
    genre1 : no_of_movies
    genre2 : no_of_movies
    genre3 : no_of_movies
    ...
  }
  reziser2 : {
    genre1 : no_of_movies
    genre2 : no_of_movies
    genre3 : no_of_movies
    ...
  }
  reziser3 : {
    genre1 : no_of_movies
    genre2 : no_of_movies
    genre3 : no_of_movies
    ...
  }
  ...
}
"""


import csv

ulaz = "mali_ulaz.csv"
izlaz = "mali_izlaz.csv"
user_input_genres = ["Drama", "Romance", "Mystery"] # ovo je korisnikov izbor koje zanrove zeli da vidi

directors = {}

with open(ulaz, "r", encoding = "utf-8-sig") as in_fajl:
  reader = csv.DictReader(in_fajl, delimiter = ";", fieldnames = ["ID", "Movie Title", "Movie Genre", "Issued", "Director", "Revenue"])

  for row in reader:
    ime_rezisera = row["Director"]
    if ime_rezisera not in directors.keys():
      directors[ime_rezisera] = {}
    
    zanrovi = row["Movie Genre"].split("|")
    for zanr in zanrovi:
      if zanr not in directors[ime_rezisera].keys():
        directors[ime_rezisera][zanr] = 1
      else:
        directors[ime_rezisera][zanr] += 1


with open(izlaz, "w", encoding = "utf-8-sig") as out_fajl:
  for director, genres in directors.items():
    izlazni_string = ""
    izlazni_string += director + ";"

    for genre, no_movies in genres.items():
      if genre in user_input_genres:
        izlazni_string += genre + " : " + str(no_movies) + "|"

    izlazni_string = izlazni_string[:-1] + "\n"
    out_fajl.write(izlazni_string)
```
