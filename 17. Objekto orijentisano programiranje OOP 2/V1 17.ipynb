Napraviti klasu koja predstavlja double ended queue odnosno red gde je moguće dodavati i skidati elemente sa početka i sa kraja reda.\
Klasa treba da omogući pozivanje len() funkcije\
Klasa treba da omogući stringovnu reprezentaciju red\
Klasa treba da implementira mogućnost dohvatanja i-tog elementa sintaksom red[i]

```python
class DoubleEndedQueue:
  def __init__(self):
    self.__red = []
  
  def __len__(self):
    return len(self.__red)

  def __str__(self):
    return f"{self.__red}"
  
  def __getitem__(self, i):
    return self.__red[i]

  def pushBack(self, e):
    self.__red.insert(0,e)
  
  def pushFront(self, e):
    self.__red.append(e)

  def popBack(self):
    self.__red.pop(0)

  def popFront(self):
    self.__red.pop(-1)
  
q = DoubleEndedQueue()
q.pushFront("1")
q.pushFront("a")
q.pushBack("xz")
print(q[1])
print(q)
q.popFront()
q.popBack()
print(q)
```

Domaci 1. Napraviti klasu koja predstavlja stek, koja se zove "Stack"\
    - Klasa treba da sadrži metode za dodavanje i skidanje elemenata sa steka\
    - Klasa treba da omogući pozivanje len() funkcije\
    - Klasa treba da omogući stringovnu reprezentaciju steka\
    - Klasa treba da sadrži docstring-ove\
    - Napisati program koji korišćenjem klase Stack proverava da li su zagrade dobro uparene 

```python
class Stack:
  """ Klasa za implementiranje rada steka.
      Maksimalna velicina steka je 512 elemenata.
      Asocirane funkcije su:
        Push, Pop, isFull, isEmpty
  """
  def __init__(self):
    self.__stack = []
    self.__fullStack = 3

  def __len__(self):
    return len(self.__stack)
  
  def __str__(self):
    return f"{self.__stack}"
  
  def push(self, e):
    """ Upisuje element na kraj steka.
        Ako je stek pun, vraca gresku.
    """
    if not self.isFull():
      self.__stack.append(e)
      return True
    else:
      print("Stack Overflow")
      return False

  def pop(self):
    """ Vraca element sa kraja steka i brise element sa steka.
        Ako je stek prazan, vraca gresku.
    """
    if not self.isEmpty():
      return self.__stack.pop(-1)
    else:
      print("Stack Underflow")
      return False
  
  def isFull(self):
    """ Vraca True ako je stek pun.
        Vraca False u suprotnom.
    """
    return len(self.__stack) == self.__fullStack
  
  def isEmpty(self):
    """ Vraca True ako je stek prazan.
        Vraca False u suprotnom.
    """
    return len(self.__stack) == 0
```

```python
def uparene_zagrade(izraz, stek):
  """ Proverava da li su zagrade u stringu "izraz"
      pravlino uparene. ko su pravilno uparene vraca
      True, ako nisu vraca False.
      Ulaz:
      - izraz - izraz nad kojim se vrsi provera
      - stek - instanca Stack-a
      Izlaz:
      - None
  """
  flagged = False
  for c in izraz:
    if c == "(":
      stek.push(c)
    elif c == ")":
      if not stek.pop():
        print("Ima visak zatvorenih zagrada.")
        flagged = True
        return False
  
  if len(stek):
    print("Imate visak otvorenih zagrada.")
    flagged = True
    return False

  if not flagged:
    print("Zagrade su dobro uparene.")
    return True



s = Stack()
tacan_izraz = "(a+b)*c-(a+c)*b"
zatvoren_izraz = "(a+b)*c)-(a+c)*b"
otvoren_izraz = "((a+b)*c-(a+c)*b"

uparene_zagrade(tacan_izraz, s)
print()
uparene_zagrade(zatvoren_izraz, s)
print()
uparene_zagrade(otvoren_izraz, s)
```
