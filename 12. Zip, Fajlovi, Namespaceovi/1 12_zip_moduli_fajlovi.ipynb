## 11.3 Zip

- **zip()** funkcija je vrlo zgodna za istovremenu iteraciju po elementima dve ili vise iterable (najcesce liste, ali mogu biti i tuple, dict) koje su **istih duzina**

- Ovo je inace nemoguce postici gnjezdenjem petlji jedne u drugu, sto em uzrokuje veliku vremensku kompleksnost programa (cini ga sporim), em ne daje zeljeni efekat

````
# Recimo da je bila trka na 100m
mesto = [1, 2, 3]
ime = ["Pera", "Zika", "Mika"]
godine = ["20", "30", "40"]

# Kako sada da iteriramo kroz ove tri liste po istim indeksima?
# Obicno kada se iterira kroz vise listi, to se radi gnjezdenjem
# for petlji

for m in mesto:
    for i in ime:
        for g in godine:
            print(f"Mesto: {m}, ime: {i}, godine: {g}")

# Medjutim, ovo pravi probleme
````

- Jedini nacin da se ovo postigne je da se iterira po indeksima, sto je u sustini ono sto zip radi, samo na elegantan nacin

````
for i in range(3):     # Liste moraju biti istih duzina
    print(f"Mesto: {mesto[i]}, ime: {ime[i]}, godine: {godine[i]}")
````

- Zip iterira kao gore po prosledjenim iterablama i vraca **tuple** objekat upakovanih clanova na odgovarajucem indeksu, svaki iz po jedne iterable

````
z = zip(mesto, ime, godine)
print(list(z))
````

````
for rezultat in zip(mesto, ime, godine):
    print(f"Mesto: {rezultat[0]}, ime: {rezultat[1]}, godine: {rezultat[2]}")
````

- Kao sto vidimo, ovo je dosta elegantnije resenje, em sintaksno, em omogucava bilo koji broj prosledjenih iterabli

- Slede neke komande koje se cesto koriste u radu sa fajlovima

````
file = open("username.csv", "r")    # Fajl je otvoren u read (r) modu (ima jos w(rite), a(ppend), r+(w+r))

sadrzaj = file.read()   # Cita ceo fajl i smesta sadrzaj u promenljivu
print(sadrzaj)
print(type(sadrzaj))

# Treba imati na umu da read() metod cita karakter po karakter, a ako mu nije
# prosledjen argument, cita sve karaktere, i to opet karakter po karakter, i
# vraca string. Glava citaca se potom nadje na kraju fajla, i sledeci poziv
# bilo kakvog citanja iz fajla nece vratiti nista
print(f"Opet citamo: {file.read()}")

file.close()    # Fajl obavezno zatvorimo kada zavrsimo s njim kako ne bi bio zauzet!
````

````
# Ako je fajl ogroman, ucitavanje u jednu promenljivu bi zauzelo
# ceo RAM. Stoga je preporucljivo, i cesto prirodno, ako je fajl
# struktuiran u vidu tabele, da se cita linija po linija
file = open("username.csv", "r")

for line in file:
    print(line, type(line))

file.close()
````

- Kraj svake linije sadrzi specijalan karakter '\n' koji oznacava prelazak u novi red

- Jos jedna specijalna sintaksa koja osigurava da ce fajl uvek biti zatvoren cak i ako zaboravite da pozovete close je:

````
with open('username.csv') as f:     # Otvara file kao f, i cim se izadje iz with konstrukta, fajl se zatvara
    read_data = f.read()
    print(read_data)

print(f.closed) # Proverimo da li je fajl zatvoren
````

- O radu sa fajlovima ima da se kaze mnogo, ali nema mnogo smisla nabrajati sve to ovde, vec je pametnije iskopati na internetu ono sto nam je potrebno u trenutku kad nam bude potrebno

1. Napisati program koji se sastoji od dve funkcije: ucitaj(ime) i prepisi(ime). Program treba da prepise sadržaj datoteke čije se ime prosleđuje kao argument fje ucitaj u datoteku čije se ime prosledjuje kao argument fje prepisi. U slučaju greške, na standardni izlaz za greške ispisati odgovarajuću poruku. Voditi racuna o pravilnom hendlovanju fajlova.

````
def ucitaj(ime):
    file = open(ime, "r")
    sadrzaj = file.read()
    file.close()
    return sadrzaj

def prepisi(ime, sadrzaj):
    file = open(ime, "w")
    file.write(sadrzaj)
    file.close()
    print("Prepiska gotova!")

sadrzaj = ucitaj("username.csv")
prepisi("prepiska.txt", sadrzaj)
````

2. Napisati program koji se sastoji iz iste funkcije ucitaj() kao u prethodnom zadatku, i funkcije prepisi_svaki_treci() koja prepisuje svaki treći karakter fajla u drugi fajl. Imena fajlova se prosledjuju kao argumenti funkcija, kao u prethodnom zadatku. U slučaju greške, na standardni izlaz ispisati odgovarajuću poruku.

````
def ucitaj(ime):
    file = open(ime, "r")
    sadrzaj = file.read()
    file.close()
    return sadrzaj

def prepisi_svaki_treci(ime, sadrzaj):
    svaki_treci = sadrzaj[::3]
    with open(ime, "w") as f:
        f.write(svaki_treci)

sadrzaj = ucitaj("prepiska.txt")
prepisi_svaki_treci("svaki_treci.txt", sadrzaj)
````
