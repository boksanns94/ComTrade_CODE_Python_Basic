# 1 Uvod u Python

## 1.1 Sta je programiranje?

* Savremeni svet nezamisliv je bez raznih elektronskih uredjaja i gedzeta koje svakodnevno koristimo

* Elektronika se do te mere zavukla u privatni zivot da nas smartphone, recimo, smatramo nezamenljivim asistentom i svojevrsnom ekstenzijom nas samih

* Prvobitni elektricni uredjaji bili su usko specijalizovani za jedan posao (elektricni motori, pumpe)

* Potencijal koji pruza struja vremenom je bivao sve elegantnije koriscen, do tacke u kojoj je razvijena jednostavna logika (ON i OFF stanja relejnih prekidaca) kojom se neki mehanicki sklop mogao kontrolisati

* Bool-ova logika (matematicki formalizam preko kog se logicke tvrdnje mogu iskazati koristeci samo dva simbola - 1 i 0, tj. stanja ON i OFF, dva diskretna naponska nivoa - "digitalna" logika) omogucila je razvoj jednostavnih AND, OR i NOT logickih kola

* 1913. Henry Sheffer - dokaz da se proizvoljno slozena logicka funkcija moze izraziti preko NAND ili NOR operatora (koji su jednostavna veza nekoliko AND, OR i NOT operatora)

* Nakon toga usledio je razvoj poluprovodnickih komponenti (tranzistora) koji su korisceni da se formiraju slozena logicka kola koja bi za neke ulazne signale davala na izlazu prethodno formiranu funkciju







Screenshot 2020-11-14 223141.png

* Prirodno, dalji put razvoja omogucio je neka logicka kola koja su bila _programabilna_, tj. bilo je moguce na neki nacin im fleksibilno menjati funkciju koju daju na izlazu, tj. _programirati_ njihov rad

* 1971. Intel je izbacio prvi zvanican procesor - _4040_

* Procesor funkcionise tako sto obradjuje instrukcije i podatke - sve date u vidu niza nula i jedinica, i na kraju izbacuje neki rezultat obrade

* Ova jednostavna obrada podataka, uz prosirenje funkcionalnosti procesora dodavanjem dodatnih hardverskih komponenti (memorije, komunikacionih kanala, raznih drugih periferija) dovela je postepenim usloznjavanjem do savremenih racunara (PC-jeva, laptopova, smartphone-a, tableta, namenskih racunara ugradjenih u automobile, avione, rutere, modeme, set-top box-ove...)

* Danas sva "pametna" elektronika funkcionise tako sto je u RAM memoriju racunara ucitan neki program (niz nula i jedinica), koji predstavlja niz instrukcija i podataka kroz koji procesor prolazi, obradjujuci ih i izbacujuci rezultate

Screenshot 2020-11-14 225217.png

* **Vazno je zapamtiti**: procesor razume samo nule i jedinice. One za njega predstavljaju instrukcije (uputstva) kako da obradi podatke koji su mu dati (takodje u vidu nula i jedinica)

* Ovo cini vrlo nezgodnim pisanje iole kompleksnijeg programa, posto ljudi nisu navikli da razmisljaju na masinskom jeziku

* Zbog ovog problema, razvijen je prvi visi nivo iznad masinskog jezika - **assembly**

* Ovaj jezik cini skup od nekoliko jednostavnih, ljudski citljivih instrukcija (poput **add**, **sub**, **mul**...) koje **asembler** prevodi u masinske instrukcije (recimo, operacija sabiranja **add** je u nekom procesoru oznacena sa 0101, tj. procesor kad vidi osmobitnu instrukciju **0101 01 10**, on zna da operande 01b = 1d i 10b = 2d treba da sabere)

* **Note**: da li je racunar 32-bitni ili 64-bitni, sto je pitanje koje se cesto cuje, prosto znaci kolike instrukcije je procesor u mogucnosti da obradi. Primer gore pokazuje da je osmobitnom instrukcijom moguce pokriti, recimo, sabiranje do nekog broja. Slicno, prosirenjem instrukcije na 32 ili 64 bita, procesoru se omogucava baratanje sa vecim podacima sto povecava i brzinu rada. Svi savremeni racunari su 64-bitni

* Sada je programeru omoguceno da napise kod u assembly-ju koji ce potom asembler prevesti u masinski jezik koji ce procesor razumeti

* Na ovaj nacin dignut je nivo apstrakcije cime je olaksan rad jer se ne mora misliti o sitnicama, vec je moguce fokusirati se na logiku programa koji treba da se izvrsava na uredjaju

* Nakon assembly-ja razvijani su tzv. _visi programski jezici_, od kojih je najznacajniji **C**

* Visi programski jezici dalje dizu nivo apstrakcije, programera lisavaju razmisljanja o hardveru, i omogucavaju mu potpuno posvecivanje razvijanju algoritma

## 1.2 Kompajler vs interpreter

* **Kompajler** je program koji prolazi kroz izvorni kod i od njega pravi izvrsni fajl (fajl na masinskom jeziku koji procesor razume)

* Sve sto je dalje potrebno je da se ovaj fajl spakuje u memoriju i procesor moze krenuti da ga obradjuje instrukciju po instrukciju

* **Interpreter** svaku pojedinacnu liniju u izvornom kodu prevodi u masinski kod koji se izvrsava, pa se potom vraca na narednu liniju koju opet prevodi u masinski kod, itd.

* Kompajler je brzi od interpretera, ali znatno kompleksniji i manje fleksibilan

## 1.4 Python 2 vs Python 3

* Python 3 je novi standard, koji je dosao uz brojne izmene i unapredjenja u odnosu na prethodnu verziju

* Brojni programi su pisani na Python-u 2 koji je dugo postojao, te se stoga cesto vidja da su na nekom sistemu (Windows-u, Linux-u) instalirana oba

## 1.11 Komentari, navodnici, tipovi podataka, input

* Svi jezici omogucavaju programeru da pored samog koda pise i komentare uz kod u samom izvornom kodu

* Komentari se ne kompajliraju niti interpretiraju - oni su samo tu kako bi programeru pomogli da se snadje u kodu i ostavlja beleske



````
# Ovo je komentar. On se ne interpretira. Pocinje tarabom na pocetku reda
# Iza tarabe moze stajati bilo sta dnoasndioasidd+saddmsl,d.a1+

a = 3 + 1   # Ovo je komentar uz naredbu levo, koja kaze da se u promenljivu a smesti rezultat 3+1

# Naredba ce biti interpretirana, a njen rezultat mozemo videti sa

print(a)    # Sada printujemo vrednost promenljive a, ne neki fiksan tekst
````

````
"""
Ovo ovde je viselinijski komentar.
On pocinje trostrukim duplim navodnicima i istim se zavrsava.
Proizvoljno mnogo linija je moguce ovako otkucati, sto se cesto
koristi kod opisa funkcija
"""
````

* U Pythonu su i jednostruki i dvostruki navodnici potpuno validni

* Fora je samo otvarati i zatvarati ono pod navodnicima istim tipom navodnika, tj. ako nesto otvorite dvostrukim, zatvorite ga dvostrukim navodnicima. Slicno za jednostruke
