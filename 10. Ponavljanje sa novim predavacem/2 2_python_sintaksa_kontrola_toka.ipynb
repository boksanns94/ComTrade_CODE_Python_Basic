## 2.2 Promenljive

* Podsetimo se jednacina u matematici
```
x + y = 7
x - y = 1
```

* **x** i **y** nazivaju se _promenljivim_ - to su simboli kojima mozemo dodeliti neke brojne vrednosti i time menjati njihovu vrednost

* Prethodni sistem dve jednacine sa dve nepoznate je potpun, sto znaci da postoji resenje sistema, i to takvo da su i **x** i **y** jednoznacni

* Iz prve jednacine imamo da je 
```
x = 7 - y
```

* Zamenom u drugu dobijamo
```
(7 - y) - y = 1
===============
7 - 2*y = 1
===============
2*y = 6
===============
y = 3
```

* Vracanjem u jednu od jednacina dobija se
```
x = 4
```

* Dakle, **x** i **y** imaju neke svoje vrednosti koje smo nasli resavanjem jednacine

* Ideja promenljivih u matematici je da se barata apstraktnim vrednostima poput slova/simbola, sve dok ne dodje trenutak da se oni zamene za konkretne vrednosti, kada citav racun dobija neki brojni smisao

* Na slican nacin, u programskim jezicima su definisane promenljive (variables) u koje se moze smestiti neka vrednost

* Kako smo rekli, uz vrednost, promenljivoj je pripisan i **tip** koji programu govori na koji nacin da tumaci tu vrednost (kao ceo ili decimalan broj, kao niz slova...)

````
a = 1               # Integer (celobrojni tip, koristi se za smestanje celih brojeva)
b = 3.14            # Floating point (decimalni tip, za smestanje brojeva sa zarezom)
c = "neka recenica" # String (tip za smestanje niza karaktera)
d = True            # Bool (logicka vrednost, postoji jos i False)
````

* Promenljiva zapravo predstavlja zamenu za memorijsku adresu racunara u kojoj je smestena sirova vrednost koja je dodeljena toj promenljivoj u programu

* Ovakva apstrakcija programera lisava rada sa memorijskim lokacijama

* Ime promenljive moze biti niz slova, brojeva i donjih crta, ali mora da pocne slovom

* Dobre prakse programiranja u Python-u https://www.python.org/dev/peps/pep-0008/

````
ovo_je_ime_promenljive = 5          # Ime se sastoji od reci i underscore-ova
ovoJeImeJosJednePromenljve = "bla"  # Camel case stil
moze_daSeRad1_i_0vAKO = False       # Ali ovo je losa praksa

# Ime promenljive treba da je opisno i da nagovestava smisao te promenljive,
# sem u trivijalnim situacijama

ocena_na_faksu = 10     # Smisleno i jasno
x = 10                  # Moze da znaci bilo sta
````

## 2.5 Razliciti nacini koriscenja funkcije print, f-string

* Kako smo videli, **print** je izuzetno fleksibilna funkcija za ispis na standardni izlaz (ekran)

````
# Mozemo odmah vrednost da ispisemo
print(5)

# Kao i rezultat neke operacije
print(1 + 5)

# Stringove takodje
print("Ovo je string")

# Dodatno, print moze primiti i vise argumenata svih gorenavedenih varijanti odvojenih zarezima
print("Rodjen sam pre", 2020 - 1996, "godine")
````

* Jedan koristan nacin da se u stringove ubaci vrednost neke promenljive je koriscenjem tzv. _f-string_-ova

* Ovo je najnoviji nacin, koriscen od Python-a 3.6.5 naovamo i drzacemo se njega

````
promenljiva = 5

s = f"Ovde cemo ubaciti vrednost {promenljiva}"

print(s)
````

## 2.6 Tipovi podataka

* Objasnili smo da tip koji je pridruzen promenljivoj sluzi programu da rastumaci na koji nacin da barata promenljivom (int, float, string, bool...)

* Ispod je data tabela sa ovim, tzv. _primitivnim_ tipovima

Screenshot 2020-11-15 124441.png

* Tip nekog podatka moguce je proveriti koriscenjem ugradjene funkcije *type*

````
print(type("neki tekst"))
print(type(4))
print(type(3.141592))
print(type(True))

nista = None
print(type(nista))   # Specijalan tip
````

* Jedan tip je moguce prevesti u drugi (ako to jezik dozvoljava, jer ne mozete recenicu, recimo, prevesti u ceo broj smisleno) koriscenjem _kastovanja_ (cast)

````
print(int(3.1415))  # Prevesce ga u ceo broj cime ce otfikariti deo iza zareza
print(float(10))    # U sustini ce dodati decimalni zarez na kraj broja (i nule iza njega)
print(float("5.5")) # I ovo je ok, ali je ovo bas nastimano da radi
print(int("Ovo vec nece moci jer nema smisla"))
````

## 2.11 Indentacija

* Python ne koristi { } kao sto to radi vecina jezika kako bi grupisao blokove naredbi

* U Pythonu se to postize indentacijom redova, sto programera tera da pise lep i uredan kod

````
if 1 > 2:
    # Ovaj blok naredbi je uvucen i sve one ce
    # se desiti ako je prvi if tacan
    print("Prvi if je tacan")
    a = 1 + 3
# Medjutim ovaj sledeci blok nije uvucen, te ne spada
# pod blok ispod if-a
print("Ovo ce se svakako desiti")
````

* U drugim C-based jezicima ovo izgleda ovako

```
if (1 > 2) {
    print(...);
    a = 1 + 3;
}

print(...);
```


* Takodje, nigde se ne koristi ; za cim ni nema potrebe ako je sve uredno pisano
