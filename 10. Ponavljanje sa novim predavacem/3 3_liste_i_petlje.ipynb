## 3.2 Liste

* Liste se koriste za smestanje niza elemenata u jednu promenljivu

* Lista je jedan od cetiri ugradjena kontejner tipa u Python-u - tipova koji sluze da drze kolekcije nekih elemenata (postoje jos tuple, set i dictionary)

* **list** je rezervisana rec!

````
ucenici = ["Pera", "Zika", "Mika"]  # 'ucenici' je lista - kolekcija tri stringa

print(ucenici)
print(f"U ovoj listi ima {len(ucenici)} elementa")
print(type(ucenici))
````

* Funkcija **len(lista)** vraca broj elemenata u listi

* Elementi u listi su uredjeni, promenljivi i dozvoljeni su duplikati

* Lista moze sadrzati elemente bilo kog tipa, ne samo primitivne tipove

* Moze sadrzati i elemente raznih tipova, ali to je pametnije ne koristiti i retko kad zaista ima potrebe za tim u praksi

````
lista_intova = [1, 3, 6, 3, 7, 2, 9, 300]

lista_boolova = [True, False, False, True]

mix_lista = ["string", False, 5, 666.666]   # Ne koristiti
````

## 3.3 Indeksiranje

* Nacin na koji pristupamo elementima liste je koriscenjem indeksa

* Indeksi u Python-u pocinju od 0

* Ovo je uradjeno iz pozadinskih implementacionih razloga brze aritmetike sa memorijskim lokacijama

Screenshot 2020-11-15 195624.png

````
lista = ['p', 'r', 'o', 'b', 'e']

print(lista)
print(lista[0])
print(lista[4])
print(lista[-3])    # Negativni indeksi znace da brojimo zdesna na levo
print(lista[5])     # Aha, izasli smo iz opsega liste
````

````
prazna_lista = []

print(prazna_lista)
print(len(prazna_lista))
````

- Moguce je indeksirati listu i tako da se uzima svaki n-ti element

````
lista = [1, 2, 3, 4, 5, 6]

# Ovakvo indeksiranje ide od pocetka do kraja
cela_lista = lista[:]

# Ovakvo indeksiranje uzima svaki drugi clan liste
svaki_drugi = lista[::2]

# Moze i od nazad
unazad = lista[::-1]

print(lista)
print(cela_lista)
print(svaki_drugi)
print(unazad)
````

## 3.4 Slajsovanje

* Cesto je potrebno iz liste izvuci vise podataka odjednom i njih smestiti u zasebnu listu

* Ovo se radi koriscenjem slicing-a

````
meseci = ["Jan", "Feb", "Mar", "Apr", "Maj", "Jun", "Jul", "Avg", "Sep", "Okt", "Nov", "Dec"]

prva_tri_meseca = meseci[0:3]   # Primetite desnu granicu slajsovanja
print(prva_tri_meseca)

poslednji_mesec_u_godini = meseci[-1]
print(poslednji_mesec_u_godini)
````

* Kada se radi slajsovanje, desna granica nije ukljucena! (ovo je kao u matematici poluotvoren skup [prvi, poslednji) )

## 3.5 Append

* U listu koja je vec formirana moze se dodati novi element koriscenjem funkcije **append**

* Ovo je zgodno kad nije unapred poznat broj elemenata koji ce se pojaviti za obradu, vec kako koji pristize, tako se pakuje u listu

````
lista = [3, 2, 9, 8]

print(lista)

# Dodaj 5 na kraj liste
lista.append(5)

print(lista)
````

## 3.6 For petlja

* **Petlje** (loops) sluze za iterativno ponavljanje bloka instrukcija

* Naredbe koje iniciraju petlje takodje se smatraju kontrolom toka programa poput if i else naredbi

* U Python-u dva tipa petlji: for i while

* For se najcesce koristi za iteriranje po odredjenoj unapred poznatoj sekvenci, a while za iteriranje sve dok se ne ispuni neki uslov

````
elementi = [1, 2, 3, 4, 5]

for element in elementi:
    print(element)
````

* Cesto se koristi **range** da se generise neki niz kroz koji se iterira

````
for i in range(1, 6):     # Slicno, desna granica nije ukljucena
    print(i)
````

* Jos jedna korisna funkcija je **enumerate**

````
likovi = ["Raja", "Gaja", "Vlaja"]

for index, lik in enumerate(likovi):
    print(index, lik)
````

## 3.7 While petlja

* While petlja vrsi iteracije kroz blok komandi sve dok je neki usput ispunjen

* Cesta pocetnicka greska: beskonacne petlje

````
i = 0

while i < 10:
    print(i)
    i += 1
````

````
# Ovo bi se zaglavilo i vecito bi ispisivalo isto dok ne ubijemo interpreter
while 1 < 2:
   print("Infinity") 
````

## 3.8 Continue i break

* Naredbe koje sluze da se iteracija preskoci ili da se prekine prevremeno

````
for i in range(1, 11):
    if i == 5 or i == 4:
        continue
    elif i == 7:
        break
    else:
        print(i)
````
